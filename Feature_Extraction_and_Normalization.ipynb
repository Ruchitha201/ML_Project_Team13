```python
import cv2
import numpy as np
import math
import csv
import glob
import pandas as pd
```

```python
from mlxtend.image import extract_face_landmarks
```

```python
def eye_aspect_ratio(eye):
	A = distance.euclidean(eye[1], eye[5])
	B = distance.euclidean(eye[2], eye[4])
	C = distance.euclidean(eye[0], eye[3])
	ear = (A + B) / (2.0 * C)
	return ear
```

```python
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar
```

```python
def circularity(eye):
    A = distance.euclidean(eye[1], eye[4])
    radius  = A/2.0
    Area = math.pi * (radius ** 2)
    p = 0
    p += distance.euclidean(eye[0], eye[1])
    p += distance.euclidean(eye[1], eye[2])
    p += distance.euclidean(eye[2], eye[3])
    p += distance.euclidean(eye[3], eye[4])
    p += distance.euclidean(eye[4], eye[5])
    p += distance.euclidean(eye[5], eye[0])
    return 4 * math.pi * Area /(p**2)
```

```python
def mouth_over_eye(eye):
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    mouth_eye = mar/ear
    return mouth_eye
```

```python
import cv2
def getFrame(sec):
    start = 180000
    vidcap.set(cv2.CAP_PROP_POS_MSEC, start+sec*1000)
    
    hasFrames,image = vidcap.read()
    #print(hasFrames)
    return hasFrames, image
```

```python
from scipy.spatial import distance
import cv2
data = []
labels = []
participant=[]
for j in [55,60]:
  for i in [0,5,10]:
    vidcap = cv2.VideoCapture('/content/drive/MyDrive/Fold5_part2' + '/' + str(j) +'/' + str(i) + '.mp4')
   
    sec = 0
    frameRate = 1
    success, image  = getFrame(sec)
    count = 0
    participant.append(j)
    while success and count < 240: 
          landmarks = extract_face_landmarks(image)
          if sum(sum(landmarks)) != 0:
              count += 1
              data.append(landmarks)
              labels.append([i])
              sec = sec + frameRate
              sec = round(sec, 2)
              success, image = getFrame(sec)
              print(count)
          else:  
              sec = sec + frameRate
              sec = round(sec, 2)
              success, image = getFrame(sec)
              print("not detected")
```

```python
import numpy as np
data = np.array(data)

labels = np.array(labels)
participant=np.array(participant)
print(participant)
```

```python
features = []
for d in range(len(data)):
  eye = data[d][36:68]
  ear = eye_aspect_ratio(eye)
  mar = mouth_aspect_ratio(eye)
  cir = circularity(eye)
  mouth_eye = mouth_over_eye(eye)
  features.append([ear, mar, cir, mouth_eye])
  print(features)
```

```python
df_alert = df[df["Y"]==0]
```

```python
df_alert_1 = df_alert.iloc[0::240, :]
df_alert_2 = df_alert.iloc[1::240, :]
df_alert_3 = df_alert.iloc[2::240, :]
```

```python
alert_first3 = [df_alert_1,df_alert_2,df_alert_3]
df_alert_first3 = pd.concat(alert_first3)
df_alert_first3 = df_alert_first3.sort_index()
```

```python
df_means = df_alert_first3.groupby("Y")["EAR","MAR","Circularity","MOE"].mean()
df_std = df_alert_first3.groupby("Y")["EAR","MAR","Circularity","MOE"].std()
```

```python
def mean_EAR(respondent):
    return df_means.loc[respondent]["EAR"]

def mean_MAR(respondent):
    return df_means.loc[respondent]["MAR"]

def mean_Circularity(respondent):
    return df_means.loc[respondent]["Circularity"]

def mean_MOE(respondent):
    return df_means.loc[respondent]["MOE"]


def std_EAR(respondent):
    return df_std.loc[respondent]["EAR"]

def std_MAR(respondent):
    return df_std.loc[respondent]["MAR"]

def std_Circularity(respondent):
    return df_std.loc[respondent]["Circularity"]

def std_MOE(respondent):
    return df_std.loc[respondent]["MOE"]
```

```python
df["EAR_mean"] = df["Y"].apply(mean_EAR)
df["MAR_mean"] = df["Y"].apply(mean_MAR)
df["Circularity_mean"] = df["Y"].apply(mean_Circularity)
df["MOE_mean"] = df["Y"].apply(mean_MOE)

df["EAR_std"] = df["Y"].apply(std_EAR)
df["MAR_std"] = df["Y"].apply(std_MAR)
df["Circularity_std"] = df["Y"].apply(std_Circularity)
df["MOE_std"] = df["Y"].apply(std_MOE)
```

```python
df["EAR_N"] = (df["EAR"]-df["EAR_mean"])/ df["EAR_std"]
df["MAR_N"] = (df["MAR"]-df["MAR_mean"])/ df["MAR_std"]
df["Circularity_N"] = (df["Circularity"]-df["Circularity_mean"])/ df["Circularity_std"]
df["MOE_N"] = (df["MOE"]-df["MOE_mean"])/ df["MOE_std"]
```

```python
df.to_csv('totalmaindata.csv',index=False)
```
